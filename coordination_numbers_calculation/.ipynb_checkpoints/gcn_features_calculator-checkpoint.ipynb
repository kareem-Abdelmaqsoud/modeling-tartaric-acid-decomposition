```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LassoLars
from sklearn.model_selection import train_test_split
from scipy.interpolate import griddata
from coordination_calculator import SurfaceAnalyzer
from utilities import chirality, half_time, predictions_df,hanging_line
from sklearn.linear_model import LinearRegression
import pickle
import ast
import os
from pymatgen.io.ase import AseAtomsAdaptor as aaa
from scipy.interpolate import LinearNDInterpolator
from itertools import chain
from collections import defaultdict
```

```python
## load the pickle files that contain the pymatgen generated slabs
path = "../slab_generation/final_gcn_slabs"
filenames = os.listdir(path)
miller_indices = []
for i in filenames:
    miller_indices.append(ast.literal_eval(os.path.splitext(i)[0]))
miller_indices = np.array(miller_indices)
len(miller_indices)
```

```python
# normalize the Miller indices to unit length
norm_miller_indices = miller_indices*np.repeat((1/np.linalg.norm(miller_indices, axis=1)).reshape(-1,1), 3, axis= 1)
# Coordinate system transformation to have (100) as the z-axis instead of (001)
# x = k, y = l, z = h (Holsclaw thesis, table #4, pg. 59)
norm_miller_indices = np.concatenate([norm_miller_indices[:,1].reshape(-1,1), norm_miller_indices[:,2].reshape(-1,1), norm_miller_indices[:,0].reshape(-1,1)], axis=1)
# Sterographic projection formula (5.5) in Holsclaw thesis
xy_prjection = np.concatenate([(norm_miller_indices[:,0]/(1+norm_miller_indices[:,2])).reshape(-1,1), (norm_miller_indices[:,1]/(1+norm_miller_indices[:,2])).reshape(-1,1)], axis =1)
```

```python
# plot the the points that represents the generated slabs
# We generated slabs that lie outside the stereographic triangle because the sections of the samples lie outside 
plt.plot(xy_prjection[:,0], xy_prjection[:,1], 'o')
#Draw the stereograhic triangle
x_values_1 = 0, 0.414214	
y_values_1 =  0, 0
plt.plot(x_values_1, y_values_1, 'k', linestyle="-", alpha = 0.5)
x_values_2 = 0, 0.366025	
y_values_2 =  0, 0.366025	
plt.plot(x_values_2, y_values_2, 'k', linestyle="-", alpha = 0.5)
point1 = [0.366025	, 0.366025]
point2 = [0, 0.414214	]
y_1,x_1 = hanging_line(point1, point2)
plt.plot(x_1,y_1,'k', linestyle="-", alpha = 0.5)
```

```python
# Since sections of the samples lie outside the triangle, here we reflect the points across the x and y coordinates 
# to get the full distribution of miller indices
# It is okay to do this reflection because the slabs that come from the relfection are the same as the original slabs
all_proj_coord = np.vstack([xy_prjection,
        np.concatenate([-xy_prjection[:,0].reshape(-1,1), xy_prjection[:,1].reshape(-1,1)], axis = 1),
        np.concatenate([-xy_prjection[:,0].reshape(-1,1), -xy_prjection[:,1].reshape(-1,1)], axis = 1),
        np.concatenate([xy_prjection[:,0].reshape(-1,1), -xy_prjection[:,1].reshape(-1,1)], axis = 1)]
)
```

```python
# use the SurfaceAnalyzer class to get the Generalized coordination numbers of unfirmally sampled slabs
from tqdm import tqdm
gcn_list = []
for i in tqdm(filenames):
    with open(os.path.join(path, i) , 'rb') as handle:
        generated_slab = pickle.load(handle)
        atoms = aaa.get_atoms(generated_slab)
        sa = SurfaceAnalyzer(atoms)
        _,_,_,_, gcn_proportions = sa.get_surface_cn_info()
        gcn_list.append(gcn_proportions)
gcn_proportions_values = np.array(gcn_list)
```

```python
# finding the unique gcn values based on the key of the gcn_proportions dictionary 
# Function to print all unique keys present in a list of dictionaries
def UniqueKeys(arr):
    # Stores the list of unique keys
    res = list(set(chain.from_iterable(sub.keys() for sub in arr)))
    return res

unique_gcns = UniqueKeys(gcn_proportions_values)
# sort the unique gcn values in ascending order
unique_gcns.sort()
```

```python
# make all the dictionaries have the same keys, putting a zero value for the coordination number types that are not present
for j in range(len(gcn_proportions_values)):
    for i in unique_gcns: 
        if (i in gcn_proportions_values[j])==False:
            gcn_proportions_values[j][i] = 0
```

```python
# make a dictionary with the coordination number proportions for all the slabs
dd = defaultdict(list)
for d in gcn_proportions_values:
    for key, value in d.items():
        dd[key].append(value)
```

```python
# the coordination number proportions for each surface, choosing the values that correspond to the unique miller indices
gcns_features = np.array(dd[unique_gcns[0]]).reshape(-1,1)
for i in unique_gcns[1:]:
    gcns_features = np.concatenate([gcns_features, np.array(dd[i]).reshape(-1,1)], axis =1)
    all_projected_coord_features = np.vstack([gcns_features,gcns_features,gcns_features,gcns_features])
```

```python
#Load Data
experimental_data_df = pd.read_csv('../experimental_dataset.csv')
#Adjust display options
pd.set_option('display.max_columns', 100)
#Display Data
experimental_data_df
```

```python
# Laod the oriented correctly Micheal data 
miller_idx = experimental_data_df[["h", 'k', "l"]].values
norm_miller_indices_p = np.concatenate([miller_idx[:,1].reshape(-1,1), miller_idx[:,2].reshape(-1,1), miller_idx[:,0].reshape(-1,1)], axis=1)
xy_p = np.concatenate([(norm_miller_indices_p[:,0]/(1+norm_miller_indices_p[:,2])).reshape(-1,1), (norm_miller_indices_p[:,1]/(1+norm_miller_indices_p[:,2])).reshape(-1,1)], axis =1)
```

```python
plt.plot(all_proj_coord[:,0], all_proj_coord[:,1], 'o')
plt.plot(xy_p[:,0], xy_p[:,1], 'o')
```

```python
def get_coordination(proj_coord):
    '''Takes in the projected XY coord that corresponds to the miller indices of each sample and 
    uses the 2d linear interpolation scheme to predict their generalized coordination Densities'''
    interp = LinearNDInterpolator(all_proj_coord, all_projected_coord_features)
    return interp(proj_coord[:,0], proj_coord[:,1])

gcn_features = get_coordination(xy_p)
```

```python
# name the columns of the dataset based on the unique GCN values 
gcn_columns = []
for i in unique_gcns:
    gcn_columns.append('GCN Density '+str(np.round(i,3)))
```

```python
# create a dataframe with the GCN features
gcn_features_df = pd.DataFrame(gcn_features)
# name the columns of the dataframe based on the GCN values
gcn_features_df.columns = gcn_columns
# combine the experimental data with the GCN features
df_combined = pd.concat([experimental_data_df, gcn_features_df], axis=1)
```

```python
# save the combined dataset to a csv file
df_combined.to_csv('../experimental_data_with_gcn_features.csv', index=False)
```

## Obtain GCN features for all the slabs inside the stereographic triangle to use for grid searching for the optimal surface structure

```python
# the miller indices that could span the whole miller indices space
miller_indices = np.array(np.meshgrid(np.linspace(0,20,21), np.linspace(0,20,21), np.linspace(0,20,21))).T.reshape(-1,3)
miller_indices = miller_indices[1:,:]
miller_indices = miller_indices.astype('int64')
norm_miller_indices = miller_indices*np.repeat((1/np.linalg.norm(miller_indices, axis=1)).reshape(-1,1), 3, axis= 1)
```

```python
xyz_100 = np.concatenate([norm_miller_indices[:,1].reshape(-1,1), norm_miller_indices[:,2].reshape(-1,1), norm_miller_indices[:,0].reshape(-1,1)], axis=1)
xyz_100f = xyz_100*np.repeat((1/np.linalg.norm(xyz_100, axis=1)).reshape(-1,1), 3, axis= 1)
xy_p = np.concatenate([(xyz_100f[:,0]/(1+xyz_100f[:,2])).reshape(-1,1), (xyz_100f[:,1]/(1+xyz_100f[:,2])).reshape(-1,1)], axis =1)
```

```python
# select points within the stereographic triangle
idx = [np.where(((xy_p[:,1]**2+xy_p[:,0]**2)**0.5 <= 0.52) & (xy_p[:,0] <= -0.73*np.cosh(xy_p[:,1]) + 1.15))]
new_xy_p = xy_p[idx[0]]
new_miller_idx = miller_indices[idx[0]]
new_norm_miller_idx = norm_miller_indices[idx[0]]
new_idx = np.where(new_xy_p[:,1] <= new_xy_p[:,0])
new_miller_idx = new_miller_idx[new_idx]
new_norm_miller_idx = new_norm_miller_idx[new_idx]
new_xy_p = new_xy_p[new_idx]
```
